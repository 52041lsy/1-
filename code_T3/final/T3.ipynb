```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.arima.model import ARIMA
```

```python
df_csv = pd.read_csv('T3_im.CSV', index_col=0, parse_dates=True)
df_csv.drop('SKU', axis=1, inplace=True)
#转置
df_csv = df_csv.T
```

```python
#将数据集划分为训练集和测试集
train = df_csv.loc['2016-06-01':'2016-06-23']   
test = df_csv.loc['2016-06-24':'2016-06-30']
```

```python
result={}
for col in train:
    #从第二列开始 提取每一列数据进行预测
    train_data = train[col]
    test_data = test[col]
    train_data_array = train_data.values.flatten()
    train_data_array = train_data_array.astype('float32')
    test_data_array = test_data.values.flatten()
    test_data_array = test_data_array.astype('float32') 
    #检验数据是否平稳
    model = ARIMA(train_data_array, order=(5, 1, 0))
    model_fit = model.fit()
    #建立ARIMA模型
    #预测

    pred = model_fit.forecast(steps=7)
    print(f'预测值：{pred}')
    print(f'真实值：{test_data.values}')
    print('-------------------------')
    result[col] = pred
```

```python
print(result)
```

```python
predictions_df = pd.DataFrame(result)
predictions_df=predictions_df.T
predictions_df.to_csv('predictions.csv', index=False)
saved_df = pd.read_csv('predictions.csv')
print(saved_df.head())
```
